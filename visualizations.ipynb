# Volcanic Activity Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load csv file
data = pd.read_csv('data/Vol_combine.csv')

# Drop rows where VEI is NaN or not numeric
data['vei'] = pd.to_numeric(data['vei'], errors='coerce')
vei_counts = data['vei'].dropna().astype(int).value_counts().sort_index()
```

```python
# Bar chart for number of volcanoes per VEI
plt.figure(figsize=(10, 6))
vei_counts.plot(kind='bar', color='maroon', edgecolor='black')

# Labels and title
plt.title('Number of Volcanoes per VEI', fontsize=16)
plt.xlabel('VEI (Volcanic Explosivity Index)', fontsize=14)
plt.ylabel('Number of Volcanoes', fontsize=14)
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
